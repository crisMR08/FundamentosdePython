## Fase 0 — Un solo diccionario (30 min)

### ¿Qué es un diccionario?

Un **diccionario** (`dict`) guarda pares **clave → valor**.  
A diferencia de una lista (acceso por posición), en un diccionario
accedemos por **nombre de campo**.

```python
# Lista: frágil — ¿qué es el índice 1?
paciente_lista = ["Germán Cerdas Valle", 28, "M", "San José"]
print(paciente_lista[1])   # 28, pero hay que saber que 1 = edad

# Diccionario: expresivo — el nombre es la documentación
paciente_dict = {
    "nombre":    "Germán Cerdas Valle",
    "edad":      28,
    "genero":    "M",
    "provincia": "San José"
}
print(paciente_dict["edad"])   # 28, y es obvio por qué
```

### Tres métodos fundamentales

```python
d = {"nombre": "Germán", "edad": 28}

d.keys()    # → dict_keys(["nombre", "edad"])
d.values()  # → dict_values(["Germán", 28])
d.items()   # → dict_items([("nombre", "Germán"), ("edad", 28)])
```

> **Conexión con S07:** `.items()` retorna pares `(clave, valor)` —
> exactamente como las tuplas que desempaquetamos con `zip()`.

> 🔮 **Predice antes de ejecutar**
>
> Si tienes `cristopher = {"nombre": "Germán", "edad": 28}`,  
> ¿qué imprime `cristopher["edad"] + 1`? ¿Y `cristopher["nombre"].upper()`?
>
> _Tu predicción:_ ___

````python
# Demostración: lista vs. diccionario con el mismo paciente
paciente_lista = ["Germán Cerdas Valle", 28, "M", "San José"]
paciente_dict  = {
    "nombre":    "Germán Cerdas Valle",
    "edad":      28,
    "genero":    "M",
    "provincia": "San José"
}

print("── Con lista ──────────────────────────────")
print(f"  {paciente_lista[0]}, {paciente_lista[1]} años")  # ¿qué son 0 y 1?

print("\n── Con diccionario ────────────────────────")
print(f"  {paciente_dict['nombre']}, {paciente_dict['edad']} años")  # autoexplicado

print("\n── Tres métodos ───────────────────────────")
print(".keys()   →", list(paciente_dict.keys())) #Devuelve las llaves 
print(".values() →", list(paciente_dict.values())) #Valores asociados a cada llave
print(".items()  →") 
for clave, valor in paciente_dict.items():
    print(f"  {clave:<12}: {valor}")
````

### Desafío 3.C — Contador con condición

### Desafío 3.D — Condición compuesta con `and` / `or`

> ⏱️ **12 minutos individual**

**Parte 1:** Encuentra pacientes con `visitas > 10` **Y** `activo == True`.  
**Parte 2:** Encuentra pacientes con enfermedad `"gripe"` **O** enfermedad `"alergia"`.

_Valores esperados:_
- Parte 1 (`and`): 5 pacientes
- Parte 2 (`or`): 4 pacientes (Bryan, Eduardo, Anthony, Angie)

````python
# Desafío 3.D — Parte 1: visitas > 10 AND activo
cont_parte_1 = 0
print("Visitas > 10 y activo == True")
for p in pacientes:
    if p['visitas'] > 10 and p['activo'] == True:
        cont_parte_1 += 1
        print(f"Nombre: {p['nombre']} | cantidad de visitas: {p['visitas']}")
        
print(f"Cantidad de pacientes que cumplen : {cont_parte_1} ")
        
        
    
print("\n")

# Desafío 3.D — Parte 2: gripe OR alergia
cont_parte_2 = 0

print ("pacientes con gripe o alergia")
for p in pacientes:
    if p['enfermedad'] == 'gripe' or p['enfermedad'] == 'alergia':
        cont_parte_2 += 1
        print(f"nombre: {p['nombre']} | enfermedad: {p['enfermedad']}")
        
print("\n")
print(f"Cantidad de pacientes que cumplen con lo esperado: {cont_parte_2}")
````
